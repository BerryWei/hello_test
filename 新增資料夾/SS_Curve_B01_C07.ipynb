```python
allStress_eq20 = []

count = 1
run = 1

for finalFiles in loadedFiles:
    listName = 'v_s15__' + str(run)
    listName = []
    
    for file in finalFiles:
        df = pd.read_csv(file, skiprows=8, sep=" ", engine='python')
        df = df.dropna(axis=1)
        try:
            df.columns = ['ATOMS id', 'type', 'x', 'y', 'z', 'v_s4', 'v_s5', 'v_s6','v_s14', 'v_s15', 'v_s16']
        except:
            print('Error.')
            print("There is nan in " + str(file))
            break
        s_df = df.astype(str)
        if sum(df['v_s15'])<-0.001 and count > 10:
            print("Negative stress, end of calculation " + str(run) + " !")
            break
        #elif s_df11["v_s15"].str.contains("nan").any() or s_df11["v_s15"].str.contains("-nan").any():
        elif any("nan" in s15 for s15 in s_df['v_s15']):
            print("Nan exists in file " + str(file))
            break
        else:
            ans = sum(df['v_s15'])
            print(ans)
            listName.append(ans)
        count += 1
    allStress_eq20.append(listName)
    print("Run " + str(run) + " finished!!!!!!!!!!!!!!!!!!!")
    count = 1
    run += 1
```

```python
allCoord_eq20 = []

count = 1
run = 1

for finalFiles in loadedFiles:
    listX = 'xSize_' + str(run)
    listX = []
    listY = 'ySize_' + str(run)
    listY = []    
    listZ = 'zSize_' + str(run)
    listZ = []
    listCoord = 'run_' + str(run)
    listCoord = []
    
    for file in finalFiles:
        df = pd.read_csv(file, skiprows=8, sep=" ", engine='python')
        df = df.dropna(axis=1)
        try:
            df.columns = ['ATOMS id', 'type', 'x', 'y', 'z', 'v_s4', 'v_s5', 'v_s6','v_s14', 'v_s15', 'v_s16']
        except:
            print('Error.')
            print("There is nan in " + str(file))
            break
        if math.isnan(df['x'][0]):
            print("There is nan in " + str(file))
            break
        elif np.sum(df['v_s15'].values)<-0.001 and count > 10:
            print("Negative stress, end of calculation " + str(run) + " !")
            break
        else:
            xLen = np.max(df['x'].values) - np.min(df['x'].values)
            yLen = np.max(df['y'].values) - np.min(df['y'].values)
            zLen = np.max(df['z'].values) - np.min(df['z'].values)
            listX.append(xLen)
            listY.append(yLen)
            listZ.append(zLen)
            print("Length of Y axis is " + str(yLen))
        count += 1
    allCoord_eq20.append([listX, listY, listZ])
    print("Run " + str(run) + " finished!!!!!!!!!!!!!!!!!!!")
    count = 1
    run += 1    
```
